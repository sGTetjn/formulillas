```python
from sympy import *
import matplotlib.pyplot as plt
import numpy as np
```

```python
def proyeccion(A):
  """
  Devuelve la matriz de proyeccion p de la matriz A
  """
   return A * (A.T * A).inv() * A.T 
def gorrito(A, b):
  """
  Devuelve la matriz x gorrito que cumple Ax = b
  """
  return (A.T * A).inv() * A.T * b
```

```python
def ajuste_recta(puntos, matrices=None):
  """
  Devuelve la recta que ajuista a los puntos  grafica dicha recta
  """
  # construye matrices A y b
  buff = [[item[0], 1] for item in puntos]
  A = Matrix((buff))
  foo = [item[1] for item in puntos]
  b = Matrix((foo))

  # Obtiene la linea ajustada
  gor = gorrito(A, b)

  # Grafica la linea
  x = np.linspace(-10, 10, 1000)
  y = gor[0] * x + gor[1]
  fig, ax = plt.subplots()

  # Grafica los puntos
  ax.plot(x, y)
  for punto in puntos:
    ax.plot(punto[0], punto[1],linestyle="None", Marker="o")
  
  # Opcionalmente muestra A y b
  if matrices:
    pprint([A, b])

  return gor
```

```python
def vector_unitario(A):
  """
  Devuelve el vector unitario de el vector A
  """
  return A / float(sqrt((A.T * A)[0]))
```

```python
def ortogonales(vectores):
  """
  Devuelve los vectores que cumplen q1⊥q2⊥q3,
  y la norma de todos ellos es igual a 1
  """
  A = vectores[0]
  B = vectores[1] - gorrito(A, vectores[1])[0] * A
  C = vectores[2] - gorrito(A, vectores[1])[0] * A - gorrito(A, vectores[2])[0] * A
  A, B, C = vector_unitario(A), vector_unitario(B), vector_unitario(C)
  return A, B, C
```

```python
def ortogonales_prueba(vectores):
  """
  Devuelve los vectores que cumplen q1⊥q2⊥...⊥qn,
  y la norma de todos ellos es igual a 1
  """
  resultado = []
  A = vectores.pop(0)
  resultado.append(A)

  # Obtiene todos los vectores ortogonales
  for index, ele in enumerate(vectores):
    buff = ele
    for indix, eli in enumerate(vectores[0:index + 1]):
      buff -= (gorrito(A, vectores[indix])[0] * A)
    resultado.append(buff)

  # Ajusta al vector unitario
  for index, ele in enumerate(resultado):
    resultado[index] = vector_unitario(ele)

  return resultado
```
